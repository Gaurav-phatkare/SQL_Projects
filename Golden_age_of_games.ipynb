#SQL Project

```python
%load_ext sql
%sql sqlite://
```

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect(':games:')
game_sales = pd.read_csv('/content/games_sales.csv')
reviews = pd.read_csv('/content/reviews.csv')

game_sales.to_sql('game_sales', conn, if_exists='replace', index=False)
reviews.to_sql('reviews', conn, if_exists='replace', index=False)
```

```python
def printQ(query):
  return pd.read_sql_query(query, conn)
```

```python
query = "SELECT * FROM game_sales;"

pd.read_sql_query(query, conn)
```

```python
query = "SELECT * FROM reviews;"

pd.read_sql_query(query, conn)
```

##1.Select all information for the top ten best-selling games & Order the results from best-selling game down to tenth best-selling

```python
query = """
SELECT * FROM game_sales
ORDER BY Total_Shipped DESC
LIMIT 10;
"""
pd.read_sql_query(query, conn)
```

##2.  Missing review scores
* Join games_sales and reviews

* Select a count of the number of games where both critic
_score and user_score are null

```python
query ="""
SELECT COUNT(*) FROM game_sales
LEFT JOIN reviews
ON game_sales.Name = reviews.Name
WHERE reviews.Critic_Score IS NULL AND reviews.User_Score IS NULL;
"""
pd.read_sql_query(query, conn)
```

```python
query = """
SELECT COUNT(*) FROM reviews
WHERE Critic_Score IS NOT NULL AND User_Score IS NOT NULL;
"""
pd.read_sql_query(query, conn)
```

##3. Years that video game critics loved
* Select release year and average critic score for each year, rounded and aliased
* Join the game_sales and reviews tables
* Group by release year
* Order the data from highest to lowest avg_critic_score and limit to 10 results



```python
query = """
SELECT g.Year , ROUND(AVG(r.Critic_Score),2) AS avg_critic_score
FROM game_sales AS g INNER JOIN reviews AS r
ON g.Name = r.Name
GROUP BY g.Year
ORDER BY ROUND(AVG(r.Critic_Score)) DESC
LIMIT 10;
"""

pd.read_sql_query(query, conn)
```
